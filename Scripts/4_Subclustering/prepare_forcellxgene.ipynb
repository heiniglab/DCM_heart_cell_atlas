```python
import scanpy as sc
import anndata
import pandas as pd
```

```python
def prep_adata_forcellxgene(adata):
    adata=anndata.AnnData(X=adata.raw.X, 
                                         obs=adata.obs, 
                                         var=adata.raw.var, 
                                         uns=adata.uns, 
                                         obsm=adata.obsm)
    adata.obs=adata.obs[['Sample', 'Patient', 'Gender', 'Region', 'Diagnosis', 'Gene', 'Clinical.dominant.mutation', 
                         'n_genes', 'n_counts', 'percent_mito', 'percent_ribo', 'scrublet_score_z', 'scrublet_score_log', 'solo_score', 
                         'cell_type', 'cell_states']]
    return adata
```

# Fibroblasts

```python
path_to_fast="/fast/AG_Huebner/"
path_in=path_to_fast+"huebner3/ANALYSES/20190926_gp_BadOyenhausen/scanpy/20210302/FB/"
```

```python
FB_all_nobackground_harmony=sc.read_h5ad(path_in + "FB_nodoublets_ANNOTATED.h5ad")
```

```python
FB_all_nobackground_harmony_cellxgene=prep_adata_forcellxgene(FB_all_nobackground_harmony)
```

```python
sc.pl.umap(FB_all_nobackground_harmony_cellxgene, 
          color=['cell_states'])
```

```python
FB_all_nobackground_harmony_cellxgene.write(path_to_fast + "huebner3/ANALYSES/20190926_gp_BadOyenhausen/scanpy/20210302/Objects_cellxgene/FB_nodoublets_CELLXGENE.h5ad")
```
